```python
with open("./fplo.out", 'r') as f:
    read = f.readlines()

read[:5]
```

```python
st1 = 'T=      0.00000      0.00000      0.00000'
st2 = 'spin 1: WF\('
orbit = []

for string in read:
    if re.match(st2, string):
        orbit.append(string)
    elif re.match(st1, string):
        f = open('output.txt', mode='a', encoding='utf-8', newline='\n')
        # f.write(orbit[-1])
        # f.write(string)
        f.close()
```
